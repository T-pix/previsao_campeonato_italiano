```python
import pandas as pd
import requests

requisitions = requests.get('https://pt.wikipedia.org/wiki/Campeonato_Italiano_de_Futebol_de_2023–24_–_Primeira_Divisão')

tables = pd.read_html(requisitions.text)
leaderboard = tables[8]
games_list = tables[11] 
display(leaderboard)
display(games_list)
#display(tables)
```

```python
name_teams = list(games_list["Casa \ Fora"])
tag_teams = list(games_list.columns)
tag_teams.pop(0)
print(name_teams)
print(tag_teams)

name_tag_teams = dict(zip(tag_teams, name_teams))
display(name_tag_teams)
```

```python
n_games_list = games_list.set_index("Casa \ Fora")
n_games_list = n_games_list.unstack().reset_index()
n_games_list = n_games_list.rename(columns={"Casa \ Fora":"Casa ","level_0":"Fora",0:"Resultado"})

def change_tags(line):
    tag = line["Fora"]
    name = name_tag_teams[tag]
    return name

n_games_list["Fora"] = n_games_list.apply(change_tags, axis=1)

n_games_list = n_games_list[n_games_list["Fora"]!= n_games_list["Casa "]]

display(n_games_list)
```

```python
n_games_list["Resultado"] = n_games_list["Resultado"].fillna("A jogar")

games_played = n_games_list[n_games_list["Resultado"].str.contains("–")]

games_not_played = n_games_list[~n_games_list["Resultado"].str.contains("–")]

display(games_played,games_not_played)
```

```python
games_played[["gols_casa","gols_fora"]] = games_played["Resultado"].str.split("–",expand=True)

games_played = games_played.drop(columns=["Resultado"])

games_played["gols_casa"] = games_played["gols_casa"].astype(int)

games_played["gols_fora"] = games_played["gols_fora"].astype(int)

display(games_played)
```

```python
avg_goals_home = games_played.groupby("Casa ").mean(numeric_only=True)
avg_goals_home = avg_goals_home.rename(columns={"gols_casa":"gols_feitos_casa",
    "gols_fora":"gols_sofridos_casa"})

avg_goals_away = games_played.groupby("Fora").mean(numeric_only=True)
avg_goals_away = avg_goals_away.rename(columns={"gols_fora":"gols_feitos_fora",
    "gols_casa":"gols_sofridos_fora"})
display(avg_goals_away,avg_goals_home)
```

```python
statistc_table = avg_goals_home.merge(avg_goals_away,left_index=True,
    right_index=True)

statistc_table = statistc_table.reset_index()

statistc_table = statistc_table.rename(columns={"Casa ": "Times"})



display(statistc_table[["Times","gols_feitos_casa","gols_feitos_fora",
    "gols_sofridos_casa","gols_sofridos_fora"]])
```

```python
from scipy.stats import poisson

def calculate_points(line):

    home_team = line['Casa ']
    away_team = line['Fora']

    lambda_home_team = statistc_table.loc[statistc_table['Times']==home_team,
        'gols_feitos_casa'].iloc[0] * statistc_table.loc[statistc_table['Times']==away_team,
            'gols_sofridos_fora'].iloc[0]

    lambda_away_team = statistc_table.loc[statistc_table['Times']==away_team,
        'gols_feitos_fora'].iloc[0] * statistc_table.loc[statistc_table['Times']==home_team,
            'gols_sofridos_casa'].iloc[0]

    prob_win_home = 0
    prob_draw = 0
    prob_win_away = 0

    for goals_home in range(8):
        for goals_away in range(8):
            prob_result = (poisson.pmf(goals_home,lambda_home_team) 
                * poisson.pmf(goals_away,lambda_away_team))
            if goals_home == goals_away:
                prob_draw += prob_result
            elif goals_home > goals_away:
                prob_win_home += prob_result
            elif goals_home < goals_away:
                prob_win_away += prob_result

    result_home = prob_win_home * 3 + prob_draw
    result_away = prob_win_away * 3 + prob_draw
    line['Pontos_casa'] = result_home
    line['Pontos_fora'] = result_away
    
    return line
```

```python
games_not_played = games_not_played.apply(calculate_points,axis=1)
display(games_not_played)
```

```python
leaderboard_update = leaderboard[['Equipe','Pts']]
leaderboard_update['Pts'] = leaderboard_update['Pts'].astype(int)

display(leaderboard_update)
```

```python
points_home = (games_not_played.groupby('Casa ').sum(numeric_only=True)
    [['Pontos_casa']])

points_away = (games_not_played.groupby('Fora').sum(numeric_only=True)
    [['Pontos_fora']])

display(points_home,points_away)
```

```python
def update_points(line):
    teams = line['Equipe']
    points = (line['Pts'] + points_home.loc[teams,'Pontos_casa'] 
              + points_away.loc[teams,'Pontos_fora'])
    return points

leaderboard_update['Pts'] = leaderboard_update.apply(update_points,axis=1)
leaderboard_update.index += 1
leaderboard_update = leaderboard_update.sort_values(by="Pts",ascending=False)
leaderboard_update = leaderboard_update.reset_index(drop=True)

display(leaderboard_update)
```
